```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
```

```python
context = """
Randy Orton is one of WWE's most accomplished superstars, known for his finishing move,
the RKO. He is a 14-time World Champion and a third-generation wrestler from the Orton family.
Orton has had iconic rivalries with superstars like John Cena, Triple H, The Undertaker,
and Bray Wyatt. He debuted in WWE in 2002 and quickly rose to fame as a member of Evolution,
alongside Triple H, Ric Flair, and Batista.

John Cena is a 16-time World Champion and one of WWE's biggest global superstars.
He has had legendary matches with Randy Orton, Edge, and AJ Styles.
Cena is known for his "Never Give Up" motto and his finishing moves, the Attitude Adjustment
and the STF submission hold.

The Undertaker is one of the most iconic WWE legends, known for his WrestleMania streak,
where he won 21 consecutive matches. He has faced Randy Orton, Brock Lesnar, Shawn Michaels,
and Triple H in historic matches.

Brock Lesnar is a dominant athlete known for his power, speed, and his finishing move,
the F5. Lesnar ended The Undertaker's undefeated WrestleMania streak in 2014.
"""
```

```python
result = qa_pipeline({
    "question": "Who is known for the RKO?",
    "context": context
})

result
```

```python
qa_pipeline({
    "question": "How many world championships has Randy Orton won?",
    "context": context
})
```
